```
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```

img = cv2.imread('dog2.jpg',cv2.IMREAD_GRAYSCALE)
img1_n = (img - img.min())/ (img.max()-img.min())
altura,largura=img.shape
h = np.zeros(256)
for x in range(0,altura):
  for y in range (0,largura):
    v = img[x,y]
    h[v] = h[v] + 1  
    
plt.plot(h)
```
